### Task 2

#### We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.

#### Task:

#### 1. fitting it to the periodic function

#### 2. plot the fit

#### Data

#### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

#### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
# import libraries
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

# define values
t_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
t_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
months = np.arange(12)

# define a function to do curve fitting
def per_fun(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

par_max, cov_max = optimize.curve_fit(per_fun, months, t_max, [20, 10, 0])
par_min, cov_min = optimize.curve_fit(per_fun, months, t_min, [-40, 20, 0])

# plot the fitting
days = np.linspace(0, 12, num=365)
plt.figure()
plt.plot(months, t_max, 'ro')
plt.plot(days, per_fun(days, *par_max), 'r-')
plt.plot(months, t_min, 'bo')
plt.plot(days, per_fun(days, *par_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()
```
